# Implementación de la derivada

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x): #Function
    return x**2
x = np.linspace(-10, 10, num=1000)

def df(f): #Derivada
    h = 0.00000000000000000000001 #Valor cercano a 0
    return (f(x+h)-f(x))/h # 'f' es la function que se aplica
y = f(x)

plt.plot(x, y, 'b')
plt.plot (x, df(f), 'r')
plt.grid()
```

```python
# Si aplicaramos 'np.cos' tambien funcionaria
plt.plot(x, y, 'b')
plt.plot (x, df(np.cos), 'r')
plt.grid()
```

## Derivada discreta

```python
def df(f):
  h=0.000001
  return (f(x+h)-f(x))/h
```

## Función lineal

$$f(x)=mx+b$$

Derivada

$$f'(x)=m$$

```python
def f(x):
  return x

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Función escalón o de Heaviside
$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Derivada

$$f´(x)=0$$

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x>=0:
      Y[idx]=1
  return Y
    
N=1000


y = H(x)

plt.plot(x, H(x), 'b')
plt.plot(x,df(H), 'r')
plt.grid()
```

## Función sigmoide

$$f(x)=\frac{1}{1-e^{-x}}$$

Derivada

$$f´(x)=f(x)(1-f(x))$$

```python
def f(x):
  return 1/(1 + np.exp(-x))
    

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Función tangente hiperbólica

$$f(x)=\frac{2}{1+e^{-2x}}-1$$

Derivada

$$f´(x)=1-f(x)^2$$

```python
def f(x):
  return np.tanh(x)
    

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```

## Función ReLU

$$R(x)=max(0,x)$$
 También se puede escribir como
$$R(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        x,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

Derivada

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def f(x):
  return np.maximum(x,0)
    

N=1000

y = f(x)

plt.plot(x, f(x), 'b')
plt.plot(x,df(f), 'r')
plt.grid()
```
